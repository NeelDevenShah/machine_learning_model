```python
import numpy as np
import pandas as pd

# If any error comes, plz refer the error docs, in it the first error
movies = pd.read_csv("C:/Users/Deven/OneDrive/Desktop/Projects/ml things/Movie Recommendation System/movies and credits/tmdb_5000_credits.csv")
credits = pd.read_csv("C:/Users/Deven/OneDrive/Desktop/Projects/ml things/Movie Recommendation System/movies and credits/tmdb_5000_movies.csv")
```

```python
# Now we will be merging both the data in the one data set only
movies = movies.merge(credits, on="title")
```

```python
print(movies.shape)
movies.head(1)
```

```python
# Now removing the columns which are not required in making the model and saving them only in the moves variable only

movies = movies[['movie_id', 'title', 'cast', 'crew', 'genres', 'keywords', 'overview']]
movies.head()
```

```python
# The ast module helps Python applications to process trees of the Python abstract syntax grammar.
import ast

# Now as the all data given in the cast, crew, genres, keyword are in the array format initially, but than it is made to be stored in the string 
# format, but now we will change them into the structured format again by the help of the bellow function:
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
# When inplace = True , the data is modified in place, which means it will return nothing and the dataframe is now updated.
movies.dropna(inplace=True)
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

```python
# As the cast have the four variables indicating about that particular cast, we have the only of them to be usefull that is name
# So, we will make an function which will in return gives an array of the only cast
# And we are only intrested in the first three cast only which are main cast, So we will make the function in such a way that it will return the cast of first three cast only

def convertCast(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies.head(2)
```

```python
# Now, we will apply the convertCast function to the cast
movies['cast'] = movies['cast'].apply(convertCast)
movies.head()
```

```python
# Now we will make an function to take director from the crew, as we are not intrested in the other crew members like camera men and etc
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
# Now, As per the recomendation given by the industry peoples, the name of the every person has the two parts, seprated by the space and by running
# the modek on such names would create the problem for the predictions, as the sam would be the surname of the more than one person, and due to the space
# beetween the names would give the wrong recomendations

# So to overcome such drawback, we will make the both the first name and last name to be combinned and for it we will make a new function as shown in the figure below:
def collapseName(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapseName)
movies['crew'] = movies['crew'].apply(collapseName)
movies['genres'] = movies['genres'].apply(collapseName)
# The collapseName can also be applied in any other things instead of the name also in which we want to remove the space between the two things
movies['keywords'] = movies['keywords'].apply(collapseName)

movies.head()
```

```python
# Now we will make all the text in one list, so that we can make an list of tags, and for it we have to convert any string data if remaning to the list form
# Here, only the overview is in the string form, all other are in the list form only

# Now lets convert the overview into the list
movies['overview'] = movies['overview'].apply(lambda x : x.split())
movies.head(3)
```

```python
# Now, we will combine overview, genres, cast, crew, keywords in the new tags columns
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['crew'] + movies['cast']
movies.head(2)
```

```python
updatedMovies = movies.drop(columns=['overview', 'keywords', 'genres', 'crew', 'cast'])
updatedMovies.head(2)
```

The problem in the above tags is that the words like corner and corners make the different featueres which are not the different features and to make this problem to be solved we have the nltk package

-->Which make the words like like, liked, liking to same one word, which helps us to make the features less and efficient

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# Example of the stem usage
print(ps.stem('loving'))
```

Now, we will apply the stem on to the our tags, but the stem can be applied to the individual elements only,So will apply the stem to the list and than will convert the list to string and than we will perform the operations

```python
# Now we will join the tags list to the string as we have the feature extraction function from the scikit learn, which will only work on the string
updatedMovies['tags'] =updatedMovies['tags'].apply(lambda x:" ".join(x))

# Conerting by help of nltk stem and also checking the change that is occuring
print(updatedMovies['tags'][0])
print("------------------")
updatedMovies['tags'] = updatedMovies['tags'].apply(stem)
print(stem(updatedMovies['tags'][0]))
# 

# Now doing the vectorization relation of the data with the updated tags
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(updatedMovies['tags']).toarray()
```

```python
# So that the each 4806 moves has the 5000 relations with each other
# The great part of the CounCountVectorizer is that, we can see the words that make the relation between the movies, as shown below:
cv.get_feature_names_out()  # OR cv.get_feature_names()
```

Now, we we have to find the distance which is the nearest or largest in the our movie vector, And for this distance the simple distamce that we find between the ending of the two vectors is not a great idea, for the efficiency and precession instead we will use the angle between them and will get the nearst one vectors by help of it, And we will perform this operation by help of the cosine_distances

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

As, we can see that the values are being not sorted in the form of the increasing values, As we have to find the 5 most nearest values i.e. the 5 movies that have the number maximum nearer to the 1

-->Also mark that the value of the same movie in the matrix would be 1 for itself, example the first movie's first column, Than the (2.2) and just like that, the diagonal would be always one

-->So, we will sort them but, In that there is an problem that the index would be shuffeled so we will not get the exact movie value as per the movie id as their index would be shuffled, So before sorting the values we will first assign the value of the index to each movie and then we will sort them

-->We will make the list of the array first As per the method shown below:
-->What does enumerate do in Python? The enumerate function in Python converts a data collection object into an enumerate object. Enumerate returns an object that contains a counter as a key for each value within an object, making items within the collection easier to access

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

Now, we will make an function that will recommend us 5 movies based on the movie that we have added or written in it

```python
def recommend(movie):
    movie_index = updatedMovies[updatedMovies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(updatedMovies.iloc[i[0]].title)
```

```python
recommend("The Avengers")
```
